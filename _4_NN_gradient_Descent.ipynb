```python
##Importing libraries for the data analytics part

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras

```

```python
data=pd.read_csv("/content/drive/MyDrive/DATASETS/insurance_data.csv")
data_org=data
data
```

```python
cols=data.columns
```

```python
# from sklearn.preprocessing import StandardScaler

# data=pd.DataFrame(StandardScaler().fit_transform(data),columns=cols)
# data
```

```python
y=data['bought_insurance']
x=data.drop('bought_insurance',axis=1)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python


x_train
```

```python
y_train
```

```python
#from keras.callbacks import EarlyStopping
```

##Implementing the artificial neural network

```python



model=keras.Sequential(

    keras.layers.Dense(1,input_shape=(2,),activation='relu')
   # keras.layers.Dense(1,activation='relu')
)


model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])



#model.fit(x_train, y_train, epochs=2000, batch_size=20, callbacks=[EarlyStopping(monitor='val_loss', patience=4)])
model.fit(x_train,y_train,epochs=100000,batch_size=50)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
result=pd.DataFrame(x_test)
result['original_value']=y_test
result['predicted_value']=np.round(y_pred)
result
```

##Evaluation methods


```python
from sklearn.metrics import precision_score,recall_score,f1_score,accuracy_score,confusion_matrix
```

```python
model.evaluate(x_test,y_test)
```
