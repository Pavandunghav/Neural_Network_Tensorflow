```python
#Importing necessary libraries 

import matplotlib.pyplot as plt
import random 
import numpy as np
```

```python
###linear function 
def linear(x):
    ''' y = f(x) It returns the input as it is'''
    return x
```

```python
x = np.linspace(-10, 10)
plt.plot(x, linear(x))
plt.axis('tight')
plt.title('Activation Function :Linear')
plt.show()
```

```python
#2]Sigmoid function 


def sigmoid(x):
    ''' It returns 1/(1+exp(-x)). where the values lies between zero and one '''

    return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function :Sigmoid')
plt.show()
```

```python
#3]ReLu

def ReLu(x):
    
    l=[]
    
    for i in x:
     if i<0:
        l.append(0)

     else:
        l.append(i)        
    
    
    return l
```

```python
x = np.linspace(-10, 10)


plt.plot(x, ReLu(x))
plt.axis('tight')
plt.title('Activation Function :ReLu')
plt.show()
```

```python
##4] tanh(x)

def tanh2(x):
    
    
    return np.tanh(x)

```

```python
x = np.linspace(-10, 10)
plt.plot(x, tanh2(x))
plt.axis('tight')
plt.title('Activation Function :tanh')
plt.show()
```

```python
##5] softmax function 

def softmax(x):
    
    return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
x = np.linspace(-10, 10)
plt.plot(x, softmax(x))
plt.axis('tight')
plt.title('Activation Function :softmax')
plt.show()
```

```python
##6]binary step 



def binarystep(x):
    
    l=[]
    
    
    
    for i in x:
     if i >=0:
        
        l.append(1)
        
        
     else:
        l.append(0)
        
        
    return l
        
        
        
        
```

```python
x = np.linspace(-10, 10)
y=binarystep(x)
plt.plot(x,y)
plt.axis('tight')
plt.title('Activation Function :binarystep')
plt.show()
```

```python
##7]LeakyReLu



def L_ReLu(x):
    
    
    l=[]
    a=0.1
    
    for i in x:
        
        if(i<0):
            l.append(a*i)
            
        else:
            l.append(i)
            
    return l
```

```python
x = np.linspace(-10, 10)
plt.plot(x, L_ReLu(x))
plt.axis('tight')
plt.title('Activation Function :leacky_relu')
plt.show()
```
